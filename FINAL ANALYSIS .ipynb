```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('FnB_hackathon_Pi.csv')
```

```python
df.head()
```

```python
df['Revenue per unit'] = df['Revenue']/df['Units sold']
```

```python
def monthChange(x):
    if x == 'JAN':
        return 1
    elif x == 'FEB':
        return 2
    elif x == 'MAR':
        return 3
    elif x == 'APR':
        return 4
    elif x == 'MAY':
        return 5
    elif x == 'JUN':
        return 6
    elif x == 'JUL':
        return 7
    elif x == 'AUG':
        return 8
    elif x == 'SEP':
        return 9
    elif x == 'OCT':
        return 10
    elif x == 'NOV':
        return 11
    elif x == 'DEC':
        return 12
```

```python
df['numMonth'] = df['MONTH'].apply(monthChange)
```

```python
dfm = df[df['Company'] == 'men']
```

```python
dfw = df[df['Company'] == 'white walkers']
```

```python
# rev/unit vs numMonth graph
```

```python
sns.barplot(y=df['Revenue per unit'], x=df['numMonth'], hue = df['Company'])
plt.xticks(fontsize=10)
plt.title('Revenue per unit per month (Company)')
```

```python
#rev/unit vs area1
```

```python
sns.barplot(y=df['Revenue per unit'], x=df['Area 1'], hue = df['Company'])
plt.xticks(rotation='vertical')
plt.title('Revenue per unit vs Area 1 (Company)')
```

```python
sns.barplot(y=df['Revenue per unit'], x=df['Area 2'], hue = df['Company'])
plt.xticks(rotation='vertical')
plt.title('Revenue per unit vs Area 2 (Company)')
```
